# 2° ETAPA:


Outra coluna existente é a University_Type (Tipo da Universidade), que informa se a Universidade é pública ou privada. Deseja-se saber a quantidade de Universidades públicas e privadas, demonstradas e explicadas graficamente. (Podem começar isolando os valores Public e Private, utilizem funções como value_counts( ), to_frame( ) e reset_index( ). Criem um dicionário para armazenar o total das universidades que são públicas e privadas, transformando-o em um DataFrame para poderem plotar os gráficos).

* Aqui vemos o tipo da universidade (publica ou privada) e o numero total de universidades de cada tipo

  # RECUPERAÇÃO QUARTA QUESTÃO:

```python
uni_pub_priv = df['University_Type'].value_counts().to_frame().reset_index()# Esta é uma linha poderosa que faz três coisas:
     # value_counts(): Conta quantas vezes "Private" e "Public" aparecem na coluna.
     # to_frame(): Transforma esse resultado (que era uma Série) em uma tabela (DataFrame).
     #reset_index(): Transforma os nomes ("Private"/"Public") em uma coluna comum, criando um novo índice numérico.

uni_pub_priv.columns = ['University_Type', 'Total'] # Ajusta os títulos das colunas para facilitar a leitura, definindo uma para a classificação e outra para o montante.
uni_pub_priv.head() # Apresenta uma prévia da tabela resultante para validar se a transformação foi bem-sucedida.
```

* Um Dicionario das informações das universidades publicas e privadas, podendo assim plotar os graficos.

```python
total_uni = df.value_counts('University_Type') # Gera um resumo quantitativo baseado na coluna de tipos.
total_geral = total_uni.sum() # Agrega todos os registros para obter o volume total da amostra (100 unidades).
total_privado = total_uni.loc['Private']  # Filtra e captura especificamente o dado numérico correspondente às instituições privadas.
total_publico = total_geral - total_privado # Utiliza o princípio da exclusão: subtraindo as privadas do montante total, define-se automaticamente a quantidade de públicas.
print(total_privado, total_publico) # Emite os resultados finais para verificação imediata no terminal.
```

* Com os dados organizados, criamos um histograma (hist) para comparar os tipos de faculdades. O gráfico mostra que a quantidade de universidades públicas e particulares é bem parecida.
Isso indica que não existe uma grande diferença entre elas e que os alunos, mesmo com menos recursos financeiros, têm boas chances de entrar em diversas instituições públicas.

```python
dados = {
    'University_Type': ['Private', 'Public'],
    'Total': [total_privado, total_publico]
} # Estrutura as informações em um objeto do tipo "Dicionário", pareando os nomes das categorias com seus respectivos totais calculados.

dados = pd.DataFrame(dados) # Transforma esse agrupamento no formato padrão do Pandas, tornando-o compatível com as ferramentas de plotagem.
sns.barplot(x='University_Type', y='Total', data=dados) # Para gerar a visualização, utilizamos a função de barras da biblioteca Seaborn (sns), configurando os seguintes parâmetros:

       # Eixo X (x): Segmenta os dados horizontalmente através da coluna de categorias, diferenciando entre instituições Públicas e Privadas.

       #Eixo Y (y): Determina a escala vertical das barras utilizando a métrica contida na coluna 'Total'.

       #Dados (data): Especifica o objeto dados (DataFrame) como a origem de onde a biblioteca deve extrair as variáveis mencionadas.
```

* Ao criarmos um gráfico de pizza, confirmamos o que vimos no histograma: existe apenas uma pequena diferença entre o número de universidades públicas e particulares.
Isso reforça que a desigualdade não é grande, mostrando que os alunos têm boas oportunidades de entrar em diversas instituições públicas, mesmo que não tenham condições financeiras.

```python
plt.pie(dados['Total'], labels = dados['Tipo da Universidade'])
```
